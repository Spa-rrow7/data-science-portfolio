## Обучение модели

```python
# Индексация категориальных признаков
indexers = [
    StringIndexer(inputCol="Gender", outputCol="GenderIndex"),
    StringIndexer(inputCol="Geography", outputCol="GeographyIndex")
]
```

```python
# Векторизация признаков
feature_cols = ["CreditScore", "GenderIndex", "Age", "Tenure", "Balance", "NumOfProducts", "IsActiveMember", "EstimatedSalary"]
assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features"
)
```

```python
# Стандартизация признаков
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")
```

```python
model_rf = RandomForestClassifier(featuresCol="scaledFeatures", labelCol="Exited", seed=25)
```

```python
pipeline = Pipeline(stages=indexers + [assembler, scaler, model_rf])
```

```python
# Разделение на train/test
train_data, test_data = data.randomSplit([0.8, 0.2], seed=25)
```

```python
# Настройка кросс-валидации
paramGrid = ParamGridBuilder() \
    .addGrid(model_rf.numTrees, [50, 100]) \
    .addGrid(model_rf.maxDepth, [5, 10]) \
    .build()
```

```python
evaluator = BinaryClassificationEvaluator(labelCol="Exited", metricName="areaUnderROC")
```

```python
crossval = CrossValidator(
    estimator=pipeline,
    estimatorParamMaps=paramGrid,
    evaluator=evaluator,
    numFolds=3
)
```

```python
cvModel = crossval.fit(train_data)
```

```python
# Оценка модели
predictions = cvModel.transform(test_data)
auc = evaluator.evaluate(predictions)

print(f"AUC on test data: {auc:.4f}")
```

Модель сильно превосходит случайные предсказания (0.5), и она на 83.36% способна корректно сортировать примеры (по вероятности ухода).

0.8336 означает, что вероятность того, что положительный пример (например, клиент, который уйдёт)
будет поставлен на более высокую позицию в предсказаниях, чем отрицательный (клиент, который не уйдёт), составляет 83.36%.
